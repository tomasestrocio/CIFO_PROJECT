<h2 style="color:#A3B18A;">5. Experimental Framework and Workflow</h2>

<p><a href="#toc" style="color:#A3B18A;">⬅ Back to Table of Contents</a></p>

<p>This section defines the framework for orchestrating all experiments. It initializes the cube moves, configures the solver and GA parameters, iterates over each combination of genetic operators, applies local and greedy improvements, and collects fitness outcomes. The results are then visualized and summarized for performance comparison.</p>

```python
if __name__ == "__main__":
    Cube.init_moves()  # This initializes all the cube moves!
    solver = RubiksCubeSolver(scramble_moves=10)    # Create a solver with 10 scramble moves
    solver.solve()                                  
```

```python
# Parameters to test
mutation_types = ['simple', 'swap', 'segment',]
crossover_types = ['one_point', 'two_point']
selection_types = ['roulette', 'tournament']

# All possible combinations of mutation, crossover, and selection types
combinacoes = list(itertools.product(mutation_types, crossover_types, selection_types))

resultados = {}  # Stores the final fitness for each combination

#  Repeat the experiment 30 times each combination
for (m_type, c_type, s_type) in combinacoes:
    print(f"\n Testing combination: Mutation={m_type}, Crossover={c_type}, Selection={s_type}")
    fitness_finais = []
    
    for i in range(30):
        Cube.init_moves()
        solver = RubiksCubeSolver(scramble_moves=10)
        ga = GeneticAlgorithm(
            scramble_cube=solver.scramble_cube,
            mutation_type=m_type,
            crossover_type=c_type,
            selection_type=s_type
        )
        solution, _, _ = ga.run()           # Run the genetic algorithm to find a solution

        # Perform local search to improve the solution
        improved = solver.local_improvement(solution)
        cube_copy = solver.scramble_cube.copy()
        cube_copy.apply_sequence(improved)
        # Perform greedy improvement to further optimize the solution
        greedy_seq, final_fit = solver.greedy_improvement(cube_copy)
        
        # Store the final fitness value after all improvements
        fitness_finais.append(final_fit)

    resultados[(m_type, c_type, s_type)] = fitness_finais
```

```python
# Plot all histograms
for key, fitnesses in resultados.items():
    plt.figure(figsize=(7, 4))
    plt.hist(fitnesses, bins=range(0, max(fitnesses)+2), edgecolor='black', align='left')
    plt.title(f"Mutation: {key[0]}, Crossover: {key[1]}, Selection: {key[2]}")
    plt.xlabel("Final Fitness")
    plt.ylabel("Frequency")
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```

```python
for key, fitnesses in resultados.items():
    average = sum(fitnesses) / len(fitnesses)  # Calculate the average fitness
    best = min(fitnesses)                       # Find the best (minimum) fitness
    print(f"{key}: Average = {average:.2f}, Best = {best}")
```

<a id="chapter6"></a>
<h2 style="color:#A3B18A;">6. Genetic Algorithm Parameters</h2>

<p><a href="#toc" style="color:#A3B18A;">⬅ Back to Table of Contents</a></p>

<p>This section outlines the key parameters used in the Genetic Algorithm, such as population size, mutation rate, number of generations, and selection strategy. These parameters are critical in guiding the evolutionary process toward optimal solutions for minimizing Rubik’s Cube moves..</p>

```python
# Parameters to test
mutation_type = 'segment'
crossover_type = 'two_point'
selection_type = 'tournament'

# Parameter ranges for testing
population_sizes = [50, 100, 150,200,300]
chromosome_lengths = [30, 40,50,60] 
crossover_rates = [0.5,0.6,0.7, 0.8] 
mutation_rates = [0.1,0.2,0.3] 
elitism_rates = [0.01, 0.10]
max_generations = [1000, 1500]

# Number of random configurations to test
n_random_configs = 30  

# Generate random configurations
random_configs = [
    (
        random.choice(population_sizes),
        random.choice(chromosome_lengths),
        random.choice(crossover_rates),
        random.choice(mutation_rates),
        random.choice(elitism_rates),
        random.choice(max_generations)
    )
    for _ in range(n_random_configs)
]

resultados = {}

# Test each random configuration
for config in random_configs:
    pop_size, chrom_len, cross_rate, mut_rate, elit_rate, max_gen = config
    print(f"\n Testing:: Pop={pop_size}, Chrom={chrom_len}, CrossRate={cross_rate}, "
          f"MutRate={mut_rate}, Elit={elit_rate}, Gens={max_gen}")
    fitness_finais = []
    
    # Run 30 trials for each configuration
    for i in range(30):
        Cube.init_moves()
        solver = RubiksCubeSolver(scramble_moves=10)
        ga = GeneticAlgorithm(
            scramble_cube=solver.scramble_cube,
            population_size=pop_size,
            chromosome_length=chrom_len,
            crossover_rate=cross_rate,
            mutation_rate=mut_rate,
            elitism_rate=elit_rate,
            max_generations=max_gen,
            mutation_type=mutation_type,
            crossover_type=crossover_type,
            selection_type=selection_type
        )
        solution, _, _ = ga.run()

        # Apply local and greedy improvements
        improved = solver.local_improvement(solution)
        cube_copy = solver.scramble_cube.copy()
        cube_copy.apply_sequence(improved)
        greedy_seq, final_fit = solver.greedy_improvement(cube_copy)
        
        fitness_finais.append(final_fit)

    resultados[config] = fitness_finais
```

```python
for key, fitnesses in resultados.items():
    media = sum(fitnesses) / len(fitnesses)     # Calculate the average fitness
    melhor = min(fitnesses)                     # Find the best (minimum) fitness
    print(f"{key}: Average = {media:.2f}, Best = {melhor}")
```

<a id="chapter7"></a>
<h2 style="color:#A3B18A;">7. Best Configuration</h2>

<p><a href="#toc" style="color:#A3B18A;">⬅ Back to Table of Contents</a></p>

```python
Cube.init_moves()

n_runs = 30
all_fitness_progress = []
final_fitness_scores = []

for i in range(n_runs):
    print(f"Run {i+1}/{n_runs}")
    solver = RubiksCubeSolver(scramble_moves=10)
    
    ga = GeneticAlgorithm(
        scramble_cube=solver.scramble_cube,
        population_size=200,
        chromosome_length=30,
        crossover_rate=0.6,
        mutation_rate=0.3,
        elitism_rate=0.01,
        max_generations=1500,
        mutation_type='segment',
        crossover_type='two_point',
        selection_type='tournament'
    )
    
    _, _, fitness_progress = ga.run()
    all_fitness_progress.append(fitness_progress)
    final_fitness_scores.append(fitness_progress[-1])  # Store the last fitness value


# Align different run lengths
max_length = max(len(run) for run in all_fitness_progress)
aligned_fitness = np.full((n_runs, max_length), np.nan)

for i, run in enumerate(all_fitness_progress):
    aligned_fitness[i, :len(run)] = run

mean_fitness_per_generation = np.nanmean(aligned_fitness, axis=0)

# Plot the average fitness evolution
plt.figure(figsize=(8, 5))
plt.plot(mean_fitness_per_generation, marker='o',color='k')
plt.xlabel("Generation")
plt.ylabel("Mean Fitness")
plt.title("Mean Fitness Evolution for 30 Runs")
plt.grid(True)
plt.show()


# Plot the distribution of final fitness values
plt.figure(figsize=(8, 5))
plt.hist(final_fitness_scores, bins=range(min(final_fitness_scores), max(final_fitness_scores)+2), edgecolor='black', align='left')
plt.xlabel("Final Fitness Score")
plt.ylabel("Frequency")
plt.title("Distribution of Final Fitness for 30 Runs")
plt.grid(True)
plt.show()
```

<a id="chapter8"></a>
<h2 style="color:#A3B18A;">8. Test Improvements Performance</h2>

<p><a href="#toc" style="color:#A3B18A;">⬅ Back to Table of Contents</a></p>


```python
Cube.init_moves()

n_runs = 30

# Store results separately for both settings
results = {
    "with_improvement": {"progress": [], "final_scores": []},
    "without_improvement": {"progress": [], "final_scores": []}
}

for use_improvement in [False, True]:
    mode = "with_improvement" if use_improvement else "without_improvement"
    print(f"\n--- Running: {mode.replace('_', ' ').title()} ---")

    for i in range(n_runs):
        solver = RubiksCubeSolver(scramble_moves=10)

        ga = GeneticAlgorithm(
            scramble_cube=solver.scramble_cube,
            population_size=200,
            chromosome_length=30,
            crossover_rate=0.6,
            mutation_rate=0.3,
            elitism_rate=0.01,
            max_generations=1500,
            mutation_type='segment',
            crossover_type='two_point',
            selection_type='tournament'
        )

        solution, _, fitness_progress = ga.run()

        if use_improvement:
            # Apply local search
            improved_solution = solver.local_improvement(solution)
            cube_copy = solver.scramble_cube.copy()
            cube_copy.apply_sequence(improved_solution)
            _, final_fit = solver.greedy_improvement(cube_copy)

        else:
            final_fit = fitness_progress[-1]

        results[mode]["progress"].append(fitness_progress)
        results[mode]["final_scores"].append(final_fit)
```

```python
# Prepare the data dictionary (if not already defined)
fitness_results = {
    'Without Improvement': results['without_improvement']['final_scores'],
    'With Improvement': results['with_improvement']['final_scores']
}

# Convert the results into a long-format DataFrame for Seaborn
data = []
for config_label, fitness_values in fitness_results.items():
    for value in fitness_values:
        data.append({
            'Final Fitness': value,
            'Condition': config_label
        })

# Create DataFrame
df = pd.DataFrame(data)

# Set theme
sns.set_theme(style="whitegrid", palette="pastel", font_scale=1.2)

# Create the plot
plt.figure(figsize=(10, 6))
ax = sns.boxplot(x='Condition', y='Final Fitness', data=df, width=0.5, linewidth=2.5, fliersize=4)

# Titles and labels
plt.title('Final Fitness Comparison: With vs Without Improvement')
plt.xlabel('Condition')
plt.ylabel('Final Fitness of Best Individual')
plt.grid(True, axis='y', linestyle='--', alpha=0.6)

plt.tight_layout()
plt.show()
```

```python
# Wilcoxon test

fitness_with_improvement = results["with_improvement"]["final_scores"]
fitness_without_improvement = results["without_improvement"]["final_scores"]

stat, p_value = wilcoxon(fitness_with_improvement, fitness_without_improvement)

print("Wilcoxon test statistic:", stat)
print("p-value:", p_value)

if p_value < 0.05:
    print("Result: Significant difference")
else:
    print("Result: No statistically significant difference")
```
